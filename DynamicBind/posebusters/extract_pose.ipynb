```python
import os
import shutil

results_dir = "./results"  # results path
docking_dir = "./dynamicbind"  # extracted docked ligand and protein files path

if not os.path.exists(docking_dir):
    os.makedirs(docking_dir)

# extract the docked ligand and protein files
for pdbid in os.listdir(results_dir):
    pdbid_path = os.path.join(results_dir, pdbid)
    if os.path.isdir(pdbid_path):
        index_folder = os.path.join(pdbid_path, "index0_idx_0")
        if os.path.exists(index_folder):
            output_pdbid_dir = os.path.join(docking_dir, pdbid)
            if not os.path.exists(output_pdbid_dir):
                os.makedirs(output_pdbid_dir)
            for filename in os.listdir(index_folder):
                if filename.startswith("rank1_ligand") or filename.startswith("rank1_receptor"):
                    source_file = os.path.join(index_folder, filename)
                    if filename.startswith("rank1_ligand"):
                        dest_file = os.path.join(output_pdbid_dir, f"{pdbid}_docked_ligand.sdf")
                    elif filename.startswith("rank1_receptor"):
                        dest_file = os.path.join(output_pdbid_dir, f"{pdbid}_docked_protein.pdb")
                    else:
                        continue
                    shutil.copy(source_file, dest_file)
                    print(f"Copied {source_file} to {dest_file}")

print("All files have been processed.")
```
